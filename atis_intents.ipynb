```python
def clean_text(text):
    text = ''.join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = ' '.join([ps.stem(word) for word in tokens if word not in stopwords])
    return text
```

```python
df_train['clean_msg'] = df_train['msg'].apply(lambda x : clean_text(x))
```

```python
df_train.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vect = CountVectorizer(analyzer=clean_text)
```

# Using TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tv = TfidfVectorizer()
```

```python
X = tv.fit_transform(df_train['clean_msg'])
```

```python
X.shape
```

```python
tv.get_feature_names()
```

```python
matrix = pd.DataFrame(X.toarray(), columns=tv.get_feature_names())
matrix.head()
```

```python
matrix['label'] = df_train['label']
matrix.head()
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC(C= 10, gamma= 'scale', kernel= 'linear')
```

```python
X = matrix.iloc[:,:-1]
y = matrix.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)
```

```python
svm.fit(X,y)
```

```python
pred = svm.predict(X_test)
```

```python
from sklearn import metrics
```

```python
def evaluation(y_test,pred):
    print('Classification Report')
    print(metrics.classification_report(y_test,pred))
    print('Accuracy Score')
    print(metrics.accuracy_score(y_test,pred))
    print('Confusion Matrix')
    print(metrics.confusion_matrix(y_test,pred))
```

```python
evaluation(y_test,pred)
```

```python
evaluation(y_test,pred)
```

```python
from sklearn.naive_bayes import MultinomialNB,BernoulliNB,GaussianNB
```

```python
def model_type(NB):
    model = NB
    model.fit(X_train,y_train)
    return model.predict(X_test)
```

```python
pred = model_type(GaussianNB())
```

```python
evaluation(y_test,pred)
```

```python
pred = model_type(MultinomialNB())
evaluation(y_test,pred)
```

```python
pred = model_type(BernoulliNB())
evaluation(y_test,pred)
```

```python
pred = model_type(BernoulliNB(alpha=0.01))
evaluation(y_test,pred)
```
